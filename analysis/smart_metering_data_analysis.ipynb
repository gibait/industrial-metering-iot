```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_folder = "/home/giba/industrial-metering-iot/data"
data_file = "industrial_metering_consumption.csv"
file_path = os.path.join(data_folder, data_file)
```

```python
RESOURCES_PRICES = {
    "iot:sensor:water": 0.46,
    "iot:sensor:gas": 0.65,
    "iot:sensor:electricity": 0.18
}
```

```python
df = pd.read_csv(file_path)
df['datetime'] = pd.to_datetime(df['t'], unit='s')
df
```

```python
fig, axs = plt.subplots(3, sharex='all')
index = 0
fig.set_size_inches(10, 8)
fig.set_dpi(90)
plt.suptitle(f"Device measured consumption over time")
plt.xlabel("Datetime")
for key, value in df.groupby(['n']):
    axs[index].set_title(key)
    axs[index].set_ylabel(f"{value.u.iloc[0]}")
    axs[index].scatter(value.datetime, value.v, label=f"{value.bn.iloc[0]}")
    axs[index].scatter(value.datetime.mean(), value.v.mean(), label="mean")
    axs[index].legend()
    index += 1
plt.show()
```

```python
fig, axs = plt.subplots(3, sharex='all')
index = 0
fig.set_size_inches(8, 6)
fig.set_dpi(90)
plt.suptitle("Sensor family measured consumption over time")
plt.xlabel("Datetime")
for key, value in df.groupby(['bn', 'n']):
    axs[index].set_ylabel(f"{value.u.iloc[0]}")
    axs[index].scatter(value.datetime, value.v, label=key[0])
    axs[index].scatter(value.datetime.mean(), value.v.mean(), label="mean", c='red', s=100)
    axs[index].legend()
    index += 1
plt.show()
```

```python
fig, axs = plt.subplots(3, sharex='all')
index = 0
fig.set_size_inches(8, 6)
fig.set_dpi(90)
plt.suptitle(f"Device monitored supply average cost in euros")
plt.xlabel("Datetime")
plt.ylabel("Price")
for key, value in df.groupby(['n']):
    axs[index].set_title(f"{value.n.iloc[0]}")
    axs[index].plot(value.datetime, value.v * value.bn.apply(lambda x: RESOURCES_PRICES[x]))
    axs[index].scatter(value.datetime.mean(), value.v.mean() * value.bn.apply(lambda x: RESOURCES_PRICES[x]).mean(), label="mean", c='red')
    axs[index].legend()
    index += 1
plt.show()
```

```python
fig, axs = plt.subplots(3, sharex='all')
index = 0
fig.set_size_inches(8, 6)
fig.set_dpi(90)
plt.suptitle(f"Sensor family monitored supply average cost in euros")
plt.xlabel("Datetime")
plt.ylabel("Price")
for key, value in df.groupby(['bn', 'n']):
    axs[index].set_title(f"{value.bn.iloc[0]}")
    axs[index].plot(value.datetime, value.v * value.bn.apply(lambda x: RESOURCES_PRICES[x]))
    axs[index].scatter(value.datetime.mean(), value.v.mean() * value.bn.apply(lambda x: RESOURCES_PRICES[x]).mean(), label="mean", c='red')
    axs[index].legend()
    index += 1
plt.show()
```
